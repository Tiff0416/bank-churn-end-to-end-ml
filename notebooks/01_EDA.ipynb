```python
df = pd.read_csv("../data/processed/bank_churners_clean_with_header.csv")

# fixing the quotes issue
df = df.applymap(lambda x: x.strip('"') if isinstance(x, str) else x)

# redo the churn flag
df["Churn_Flag"] = (df["Attrition_Flag"] == "Attrited Customer").astype(int)

df["Churn_Flag"].value_counts()

import seaborn as sns
import matplotlib.pyplot as plt

# Visualizing churn rates across categorical features
cat_cols = [
    "Gender",
    "Education_Level",
    "Marital_Status",
    "Income_Category",
    "Card_Category"
]

for col in cat_cols:
    plt.figure(figsize=(6, 4))

    churn_rate = (
        df.groupby(col)["Churn_Flag"]
          .mean()
          .sort_values(ascending=False)
    )

    sns.barplot(x=churn_rate.index, y=churn_rate.values)
    plt.title(f"Churn Rate by {col}", fontsize=14)
    plt.ylabel("Churn Rate")
    plt.xlabel(col)
    plt.xticks(rotation=45, ha="right")
    plt.ylim(0, churn_rate.max() + 0.02)
    plt.tight_layout()
    plt.show()

df["Has_Dependents"] = (df["Dependent_count"] > 0).astype(int)

has_dep_churn = (
    df.groupby("Has_Dependents")["Churn_Flag"]
      .mean()
      .sort_index()
)

plt.figure(figsize=(5, 4))
sns.barplot(x=["No Dependents", "Has Dependents"],
            y=has_dep_churn.values)
plt.title("Churn Rate: With vs Without Dependents")
plt.ylabel("Churn Rate")
plt.tight_layout()
plt.show()


plt.figure(figsize=(6, 4))
sns.boxplot(x="Attrition_Flag", y="Months_on_book", data=df)
plt.title("Customer Tenure (Months_on_book) by Attrition Status")
plt.xlabel("Attrition_Flag")
plt.ylabel("Months_on_book")
plt.tight_layout()
plt.show()

import numpy as np

max_m = df["Months_on_book"].max()

bins = [0, 24, 36, 48, 60, 100]
labels = ["<2 years", "2–3 years", "3–4 years", "4–5 years", "5+ years"]

df["Tenure_bin"] = pd.cut(df["Months_on_book"],
                          bins=bins,
                          labels=labels,
                          right=False)

tenure_churn = (
    df.groupby("Tenure_bin")["Churn_Flag"]
      .mean()
      .reindex(labels)
)

plt.figure(figsize=(6, 4))
sns.barplot(x=tenure_churn.index, y=tenure_churn.values)
plt.title("Churn Rate by Tenure Bin")
plt.xlabel("Tenure")
plt.ylabel("Churn Rate")
plt.ylim(0, tenure_churn.max() + 0.02)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Our exploratory analysis shows that customer churn is influenced by a combination of demographic, product, household, and tenure characteristics:

- Gender: Female customers have a slightly higher churn rate than male customers, but the difference is modest, suggesting gender is not a primary driver of attrition.

- Education Level: Customers with higher education (Doctorate and Post-Graduate) show noticeably higher churn rates. This may indicate higher expectations for product value or a greater likelihood of switching to competing credit card offers.

- Marital Status: Married customers exhibit the lowest churn rate and appear to be the most stable segment. Single and “Unknown” customers have comparatively higher attrition rates, reflecting lower long-term engagement.

- Income Category: Churn displays a U-shaped pattern. Both low-income and high-income customers churn at higher rates, while middle-income customers are the most stable. This suggests two distinct sources of risk: financial strain at the low end and active card-shopping behavior at the high end.

- Card Category: Platinum cardholders show the highest churn rate, whereas Silver and Blue customers churn the least. This unusual pattern may reflect a small Platinum sample size or a tendency for high-tier users to switch providers when dissatisfied.

- Dependent Count: Customers with dependents churn only slightly more than those without dependents. The small difference indicates that household size is not a major determinant of churn.

- Tenure (Months_on_book): Churn increases slightly with tenure—customers with more than three years of card ownership churn at the highest rate. However, the overall differences are mild, suggesting tenure is a secondary factor compared with behavioral features.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 想看的「行為變數」清單（含 Credit_Limit）
behavior_vars = [
    "Total_Trans_Ct",
    "Total_Trans_Amt",
    "Months_Inactive_12_mon",
    "Contacts_Count_12_mon",
    "Total_Relationship_Count",
    "Avg_Utilization_Ratio",
    "Credit_Limit"
]

for var in behavior_vars:
    plt.figure(figsize=(7, 4))
    sns.boxplot(data=df, x="Attrition_Flag", y=var)
    plt.title(f"{var} by Attrition Status")
    plt.xlabel("Attrition_Flag")  
    plt.tight_layout()
    plt.show()
```

Across all behavioral indicators, attrited customers consistently show a weaker engagement profile. They make fewer transactions and spend substantially less compared to retained customers, signaling declining activity well before attrition occurs. Attrited customers also record more inactive months and reach out to customer service more frequently, reflecting increasing friction or dissatisfaction with the bank. In addition, churned customers tend to hold fewer product relationships, have lower credit utilization, and generally operate with lower effective credit use despite having similar credit limits. Collectively, these patterns indicate that decreasing engagement, reduced product depth, and rising service-related friction are the strongest behavioral predictors of customer attrition.

```python
import seaborn as sns
import matplotlib.pyplot as plt

change_vars = [
    "Total_Amt_Chng_Q4_Q1",
    "Total_Ct_Chng_Q4_Q1"
]

for var in change_vars:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=df, x="Attrition_Flag", y=var)
    plt.title(f"{var} by Attrition Status")
    plt.xlabel("Attrition_Flag")
    plt.ylabel(var)
    plt.tight_layout()
    plt.show()
```

Churned customers show noticeably lower Q4-to-Q1 changes in transaction amount. Their median spending ratio is significantly lower than that of existing customers, indicating that spending declines substantially prior to attrition. Meanwhile, existing customers exhibit higher and more dispersed change ratios, including many cases where Q4 spending exceeds Q1, signaling healthy usage. This downward shift in spending is a strong behavioral predictor of churn.

Transaction count change exhibits a similar pattern: churned customers have substantially lower Q4-to-Q1 ratios, reflecting a sharp decline in purchasing frequency. Existing customers show higher ratios with more variance, including numerous cases where transaction counts grew over time. This reinforces that decreasing engagement—both in dollar amount and in transaction frequency—is a strong early indicator of future attrition.

留下來的客戶：Q4 的消費金額比 Q1 更高或差不多 → 消費穩定/成長

```python
import seaborn as sns
import matplotlib.pyplot as plt

df["Churn_Flag"] = (df["Attrition_Flag"] == "Attrited Customer").astype(int)

df["Churn_Label"] = df["Churn_Flag"].map({0: "Existing Customer", 1: "Attrited Customer"})

# Transactions Count vs Amount
plt.figure(figsize=(6, 5))
sns.scatterplot(
    data=df,
    x="Total_Trans_Ct",
    y="Total_Trans_Amt",
    hue="Churn_Label",
    alpha=0.6
)
plt.title("Total Transactions vs Total Amount")
plt.xlabel("Total_Trans_Ct (12-month count)")
plt.ylabel("Total_Trans_Amt (12-month amount)")
plt.tight_layout()
plt.show()

# Transactions Count vs Inactive Months
plt.figure(figsize=(6, 5))
sns.scatterplot(
    data=df,
    x="Total_Trans_Ct",
    y="Months_Inactive_12_mon",
    hue="Churn_Label",
    alpha=0.6
)
plt.title("Transactions vs Inactive Months")
plt.xlabel("Total_Trans_Ct (12-month count)")
plt.ylabel("Months_Inactive_12_mon")
plt.tight_layout()
plt.show()

# Transactions Amount vs Utilization Ratio
plt.figure(figsize=(6, 5))
sns.scatterplot(
    data=df,
    x="Total_Trans_Amt",
    y="Avg_Utilization_Ratio",
    hue="Churn_Label",
    alpha=0.6
)
plt.title("Total Amount vs Utilization Ratio")
plt.xlabel("Total_Trans_Amt (12-month amount)")
plt.ylabel("Avg_Utilization_Ratio")
plt.tight_layout()
plt.show()

# Activity Level vs Activity Change(times)
plt.figure(figsize=(6, 5))
sns.scatterplot(
    data=df,
    x="Total_Trans_Ct",
    y="Total_Ct_Chng_Q4_Q1",
    hue="Churn_Label",
    alpha=0.6
)
plt.title("Transaction Count vs Count Change (Q4/Q1)")
plt.xlabel("Total_Trans_Ct (12-month count)")
plt.ylabel("Total_Ct_Chng_Q4_Q1")
plt.tight_layout()
plt.show()

# Amount Level vs Amount Change
plt.figure(figsize=(6, 5))
sns.scatterplot(
    data=df,
    x="Total_Trans_Amt",
    y="Total_Amt_Chng_Q4_Q1",
    hue="Churn_Label",
    alpha=0.6
)
plt.title("Transaction Amount vs Amount Change (Q4/Q1)")
plt.xlabel("Total_Trans_Amt (12-month amount)")
plt.ylabel("Total_Amt_Chng_Q4_Q1")
plt.tight_layout()
plt.show()

# Product Depth vs Transactions
plt.figure(figsize=(6, 5))
sns.scatterplot(
    data=df,
    x="Total_Relationship_Count",
    y="Total_Trans_Ct",
    hue="Churn_Label",
    alpha=0.6
)
plt.title("Product Relationships vs Transactions")
plt.xlabel("Total_Relationship_Count")
plt.ylabel("Total_Trans_Ct (12-month count)")
plt.tight_layout()
plt.show()

```

The scatter plots reveal clear behavioral segmentation patterns that strongly differentiate churned from retained customers. Churned customers consistently cluster in the low-engagement region—characterized by fewer transactions, lower spending, long inactive periods, and declining activity trends. They also exhibit lower credit utilization, suggesting reduced card reliance or migration to other cards. In contrast, retained customers are concentrated in high-activity clusters with stable or growing transaction behavior and higher utilization. Additionally, multi-product customers show much stronger retention, reinforcing the role of relationship depth. Overall, customer churn in this dataset is driven by a combination of declining activity, inactivity buildup, and weak product engagement, rather than demographic factors.

```python
import numpy as np
import matplotlib.pyplot as plt

# right skewed columns
skew_cols = [
    "Total_Trans_Amt",
    "Total_Trans_Ct",
    "Total_Revolving_Bal",
    "Avg_Open_To_Buy",
    "Credit_Limit",
    "Total_Amt_Chng_Q4_Q1",
    "Total_Ct_Chng_Q4_Q1",
    "Avg_Utilization_Ratio"
]

# Plot original vs log1p transformed distributions
for col in skew_cols:
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))

    # original distribution
    ax[0].hist(df[col], bins=40, color="steelblue")
    ax[0].set_title(f"Original {col}")

    # log1p transformed distribution
    ax[1].hist(np.log1p(df[col]), bins=40, color="orange")
    ax[1].set_title(f"log1p({col})")

    plt.tight_layout()
    plt.show()
```

| Feature                 | 建議處理                               | 為什麼                                |
| ----------------------- | ---------------------------------- | ---------------------------------- |
| **Total_Trans_Amt**     | ✔ log transform                    | 極度右偏，金額差異大，log 後更正常                |
| **Avg_Open_To_Buy**     | ✔ log transform                    | Long tail，到 34K，log 後分佈漂亮          |
| **Credit_Limit**        | ✔ log transform                    | 跟上面一樣，額度長尾現象很強                     |
| **Total_Revolving_Bal** | ✔ log transform｜✔ 建議加 binary（是否=0） | RevoBal 大部分是 0 → log 只解決後半段，需二元化補強 |


| Feature                  | 建議處理         | 為什麼                   |
| ------------------------ | ------------ | --------------------- |
| **Total_Trans_Ct**       | optional log | 偏度不強，但 log 可讓線性模型收斂更好 |
| **Total_Amt_Chng_Q4_Q1** | optional log | 已經接近正態，log 差異不大       |
| **Total_Ct_Chng_Q4_Q1**  | optional log | 同上；樞紐變數，log 不影響太多     |


| Feature                      | 建議處理                                   | 原因                              |
| ---------------------------- | -------------------------------------- | ------------------------------- |
| **Avg_Utilization_Ratio**    | ✔ 分箱（0, 0–0.1, 0.1–0.3, 0.3–0.6, >0.6） | 行為模式非線性：低/中/高 使用率差異很大           |
| **Total_Revolving_Bal**      | ✔ 變成 0 vs non-zero                     | 70% 是 0，二分類比較有 predictive power |
| **Months_Inactive_12_mon**   | ✔ binned（0–1、2–3、4–6）                  | 分佈序列化 but 離散值表示行為（非連續）          |
| **Contacts_Count_12_mon**    | ✔ 保留原值 + optional binning              | count 數量小，分箱反而解釋性強              |
| **Total_Relationship_Count** | ✔ 不做 log｜✔ categorical encoding        | 數值本身 = "產品數"，不是連續變量             |
| **Dependent_count**          | ✔ 不做 log｜✔ 直接保留                        | 小範圍 discrete values（0–5）不適合 log |


| Feature                          | 為什麼                               |
| -------------------------------- | --------------------------------- |
